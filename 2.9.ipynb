```julia
using DataFrames
using StatsBase
using Compose
using Tk
using Gadfly
```

```julia
df = readtable("Auto.txt")
head(df)
```

(a) Quantitative predictors: mpg, cylidners, displacment, horsepower, weight, acceleration, year.
Qualitative predictors: name, origin
(b,c) Summary stats are shown below.

```julia
for var in names(df)
    if (var == :origin) | (var == :name)
        continue
    end
    x = dropna(df[var]) # ignore NA
    @printf("%s\n\tRange: %f to %f\n\tMean: %f\n\tStandard deviation: %f\n", 
        var, 
        minimum(x), 
        maximum(x), 
        mean(x),
        std(x))
end
```

(d) Removing the 10-85th observation...

```julia
for var in names(df)
    if (var == :origin) | (var == :name)
        continue
    end
    x = dropna(df[[1:9,86:end],var]) # ignore NA
    @printf("%s\n\tRange: %f to %f\n\tMean: %f\n\tStandard deviation: %f\n", 
        var, 
        minimum(x), 
        maximum(x), 
        mean(x),
        std(x))
end
```
